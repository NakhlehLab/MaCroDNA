```python
# read rna data
import os
rna_raw = []
base_path = "/home/xiru/Dropbox/Research/Datasets/scTrio_seq2/CNV_Bian/GSE97693_RAW/"
file_path = os.listdir(base_path)
for fp in file_path:
    if os.path.isfile(os.path.join(base_path,fp)):
        fp_name = fp.rstrip(".txt.gz")
        #fp_name = fp_name.rstrip(".FPK")
        fp_name = fp_name.rstrip(".TPM")
        fp_name = fp_name.rstrip(".FPKM")
        fp_name = fp_name.replace("_scTrioSeq2Rna","")
        fp_name = fp_name.split("_")[1:]
        fp_name = "_".join(fp_name)
        tmp_rna = pd.read_csv(os.path.join(base_path,fp),sep="\t", names=[fp_name],index_col=0)
        rna_raw.append(tmp_rna)
```

```python
for i in range(len(rna_raw)):
    if i == 0:
        rna_df = rna_raw[0]
    else:
        rna_df = pd.concat([rna_df, rna_raw[i]], axis=1)
rna_df
```

```python
cnv_name = list(cnv_df.columns)[3:]
rna_name = list(rna_df.columns)
rna_cnv_overlap = list(set(cnv_name).intersection(rna_name))
len(cnv_name)
```

```python
cnv_df
```

```python
rna_cnv_dict={}
ppl = list(set(x[:5] for x in rna_cnv_overlap))
for p in ppl:
    rna_cnv_dict[p] = []

for rc in rna_cnv_overlap:
    rna_cnv_dict[rc[:5]].append(rc)
```

```python
for k in rna_cnv_dict.keys():
    print("%s has %d cells"%(k, len(rna_cnv_dict[k])))
```

```python
# process all cnv data
chr_name = []
start = []
end = []
copy_number = []
clone = []
chr_len = len(list(cnv_df["CHR"]))
for c_name in cnv_name:
    chr_name.extend(list(cnv_df["CHR"]))
    start.extend(list(cnv_df["START"]))
    end.extend(list(cnv_df["END"]))
    clone.extend([c_name]*chr_len)
    c_number = cnv_df[c_name].to_numpy()*2
    c_number = np.around(c_number).astype(int)
    copy_number.extend(c_number)

cnv_final_df = pd.DataFrame(list(zip(chr_name,start,end,copy_number,clone)), columns=["chr","start","end","copy_number","clone"]) 
# cnv_final_df.to_csv("clonealign_cnv.csv", header=True)
cnv_final_df
```

```python
# delete sex chr

cnv_final_df.to_csv("../Data/all_cnv.csv", header=True)
cnv_final_df_no_X = cnv_final_df[cnv_final_df["chr"]!= "chrX"]
cnv_final_df_no_X.to_csv("../Data/all_cnv_no_X.csv", header=True)
```

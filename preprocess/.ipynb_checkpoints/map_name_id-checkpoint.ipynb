```python
import pandas as pd
import numpy as np
```

```python
rna_data = pd.read_csv("all_gene_id_name.csv", index_col=0)
rna_data
```

```python
rna_data = rna_data.groupby(["gene_id"]).sum()
rna_data
```

```python
rna_data.index = list(rna_data.index)
rna_data
```

```python
rna_data.to_csv("all_rna_processed.csv", header=True,index=True)
```

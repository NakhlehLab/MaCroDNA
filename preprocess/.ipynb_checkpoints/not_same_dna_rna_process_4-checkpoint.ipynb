```python
# process patient

def extract_patient(cell_df):
#     get all cells
    cell_list = list(cell_df.columns)
#     get all patients
    patient_list = list(set([x[:5] for x in cell_list]))
    patient_list = [x for x in patient_list if x != "CRC01" and x != "CRC02"]
    
#     get dictionary, patient: patient_cell 
    dict_result = {}
    for p in patient_list:
        dict_result[p]=[]
    for c in cell_list:
        if c[:5] in patient_list:
            dict_result[c[:5]].append(c)
    
    
#     get rna dna dictionary, patient: patient rna/dna df
    na_result = {}
    for p in patient_list:
        tmp_cell = dict_result[p]
        na_result[p] = cell_df[tmp_cell]
    
    return na_result

rna_patient_dict = extract_patient(all_rna_processed)
dna_patient_dict = extract_patient(all_dna_processed)
dna_patient_dict_no_X = extract_patient(all_dna_processed_no_X)
```

```python
# find same genes in rna and dna
# and delete genes has zero in all rna cells or all dna cells

def non_zero_var_gene(cell_df):
    
#     cell_df["var"] = cell_df.var(axis=1)
#     cell_df_select = cell_df[cell_df["var"] != 0]
#     gene = list(cell_df_select.index)
    cell_var = list(cell_df.var(axis=1))
    gene_all = list(cell_df.index)
    gene = [gene_all[i] for i in range(len(cell_var)) if cell_var[i] != 0]
    
    return gene

def non_zero_gene(cell_df, percentage = 0.01):
    cell_amount = len(cell_df.columns)
    sum_thresh = cell_amount * percentage
    df_mask = copy.deepcopy(cell_df)
    df_mask = df_mask.mask(df_mask > 0, 1)
    df_mask = df_mask[(df_mask.sum(axis = 1) > sum_thresh)]
    keep_gene = list(df_mask.index)
    return keep_gene
    
    
    
def find_genes(dna_df, rna_df, remove_var0=False, remove_0=True):
    rna_gene = list(rna_df.index)
    dna_gene = list(dna_df.index)
    same_gene = set(rna_gene).intersection(dna_gene)
    
#     remove 0
    rna_gene_filter_0 = non_zero_gene(rna_df)
    dna_gene_filter_0 = non_zero_gene(dna_df)
    result_gene = same_gene.intersection(rna_gene_filter_0)
    result_gene = result_gene.intersection(dna_gene_filter_0)
    
    if remove_var0:
        rna_gene_nonzero = non_zero_var_gene(rna_df)
        dna_gene_nonzero = non_zero_var_gene(dna_df)
        result_gene = result_gene.intersection(rna_gene_nonzero)
        result_gene = result_gene.intersection(dna_gene_nonzero)
    else:
        result_gene = result_gene
#     return result_gene
    return dna_df.loc[result_gene,:], rna_df.loc[result_gene,:]




union_patient = list(set(rna_patient_dict.keys()).intersection(dna_patient_dict.keys()))

dna_patient_dict_processed = {}
dna_patient_dict_processed_no_X = {}
rna_patient_dict_processed = {}
rna_patient_dict_processed_no_X = {}

# new folder
wX_data_dir = "../Data/same_gene_with_X_no_0/"
if not os.path.exists(wX_data_dir):
    os.makedirs(wX_data_dir)

# save data without removing var 0 for all genes, and use this for the following clustering
for p in union_patient:
    dna_patient_dict_processed[p],rna_patient_dict_processed[p]=\
    find_genes(dna_patient_dict[p],rna_patient_dict[p], remove_var0=False)
    print("With X dna for ",p,": ",dna_patient_dict_processed[p].shape)
    print("With X rna for ",p,": ",rna_patient_dict_processed[p].shape)
    
    dna_patient_dict_processed[p].to_csv(wX_data_dir+str(p).lower()+"_dna_raw.csv", header=True, index=True)
    rna_patient_dict_processed[p].to_csv(wX_data_dir+str(p).lower()+"_rna_raw.csv", header=True, index=True)
    

    
# save data without var0 genes for data with no X
noX_data_dir = "../Data/clonealign_preprocess/"
if not os.path.exists(noX_data_dir):
    os.makedirs(noX_data_dir)

for p in union_patient:
    
    dna_patient_dict_processed_no_X[p],rna_patient_dict_processed_no_X[p]=\
    find_genes(dna_patient_dict_no_X[p],rna_patient_dict[p], remove_var0=True)
    
    print("No X dna for ",p,": ",dna_patient_dict_processed_no_X[p].shape)
    print("No X rna for ",p,": ",rna_patient_dict_processed_no_X[p].shape)
    
    dna_patient_dict_processed_no_X[p].to_csv(noX_data_dir+str(p).lower()+"_dna_raw.csv", header=True, index=True)
    rna_patient_dict_processed_no_X[p].to_csv(noX_data_dir+str(p).lower()+"_rna_raw.csv", header=True, index=True)
```

```python
# normalise data
# normalise data
# normalise data

from scanpy.pp import log1p

def save_data(df_list,method, dna=True):
    patient_ls = list(df_list.keys())
    if dna:
        data_type = "dna"
    else:
        data_type = "rna"
        
    data_dir = "../Data/same_gene_with_X_no_0/"
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)
        
    
    for i in range(len(patient_ls)):
        tmp_path = data_dir+str(patient_ls[i]).lower()+"_"+data_type+"_"+method+".csv"
        df_list[patient_ls[i]].to_csv(tmp_path,header=True, index=True)
        
# log transform
def log_transform(df_list):
    result_df_list={}
    for p in df_list.keys():
        tmp_df = df_list[p]
        result_df = pd.DataFrame(log1p(tmp_df.to_numpy().T).T, 
                                 index=list(tmp_df.index),columns=list(tmp_df.columns))
        result_df_list[p] = result_df
    return result_df_list
```

```python
# transform and save data

# dna data
dna_log=log_transform(dna_patient_dict_processed)
save_data(dna_log,"log", dna=True)


# rna data
rna_log=log_transform(rna_patient_dict_processed)
save_data(rna_log,"log", dna=False)
```

```python
def cluster_cnv(dna_cluster):
    cluster_id = list(set(list(dna_cluster["clone"])))
    for i in range(len(cluster_id)):
        iid = cluster_id[i]
        tmp_cluster_id = list(dna_cluster[dna_cluster["clone"] == iid]["cell"])
        tmp_cluster_df = dna[tmp_cluster_id]
        tmp_cluster_mean = tmp_cluster_df.mean(axis=1)
        
        if i == 0:
            cluster_cnv_df = tmp_cluster_mean.to_frame(name="clone"+str(iid))
            
        else:
            cluster_cnv_df["clone"+str(iid)] = tmp_cluster_mean
            
    return cluster_cnv_df

def patient_cluster_cnv(dna_cluster, dna_df):
    cluster_id = list(set(list(dna_cluster["clone"])))
    for i in range(len(cluster_id)):
        iid = cluster_id[i]
        tmp_cluster_id = list(dna_cluster[dna_cluster["clone"] == iid]["cell"])
        tmp_cluster_df = dna_df[tmp_cluster_id]
        tmp_cluster_mean = tmp_cluster_df.mean(axis=1)
        
        if i == 0:
            cluster_cnv_df = tmp_cluster_mean.to_frame(name="clone"+str(iid))
            
        else:
            cluster_cnv_df["clone"+str(iid)] = tmp_cluster_mean
            
    return cluster_cnv_df
```

```python
# agg cluster
from sklearn.cluster import AgglomerativeClustering
cluster_num = [4]

def agg_cluster(nc,tmp_df):
    tmp_result = AgglomerativeClustering(n_clusters=nc,affinity="euclidean",linkage="ward").fit_predict(tmp_df.to_numpy().T)
    return tmp_result

def generate_agg_cluster_file(c_n, patient_dna, patient_name, folder_path):
    tmp_cluster = agg_cluster(c_n, patient_dna)
    tmp_cell_id = list(patient_dna.columns)
    tmp_cluster_df = pd.DataFrame(tmp_cluster, columns=["clone"])
    tmp_cluster_df["cell"] = tmp_cell_id
    tmp_cluster_cnv_df = patient_cluster_cnv(tmp_cluster_df, patient_dna)
    assert len(patient_dna.index) == len(tmp_cluster_cnv_df.index)
    tmp_cluster_df.to_csv(folder_path+"dna_cluster_"+patient_name+"_"+str(c_n)+".csv", header=True)
    tmp_cluster_cnv_df.to_csv(folder_path+"dna_cluster_"+patient_name+"_"+str(c_n)+"_cnv.csv", header=True)

def final_agg(df_dict, cluster_num, norm_method, folder_dir="../Data/same_gene_with_X_no_0/Clusters/"):
    
    if not os.path.exists(folder_dir):
        os.makedirs(folder_dir)
    
    if norm_method != "":
        norm_method = "_"+norm_method 
    agg_path = folder_dir+"agglomerative"+norm_method+"/"
    if not os.path.exists(agg_path):
        os.makedirs(agg_path)

    for p in df_dict.keys():
        tmp_dna_df = df_dict[p]
        tmp_patient_id = p.lower()
        for c_n in cluster_num:
            generate_agg_cluster_file(c_n, tmp_dna_df, tmp_patient_id, agg_path)
            
            
final_agg(dna_patient_dict_processed, cluster_num, "raw")
final_agg(dna_log, cluster_num, "log")
# final_agg(dna_scinet, cluster_num, "scinet", data_dir)
```

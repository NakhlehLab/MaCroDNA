```python
import numpy as np
import pandas as pd
```

```python
dna = pd.read_csv("all_cnv_processed.csv",index_col=0)
rna = pd.read_csv("all_rna_processed.csv",index_col=0)
```

```python
dna_gene = list(dna.index)
rna_gene = list(rna.index)
same_gene = list(set(dna_gene).intersection(rna_gene))
same_gene
```

```python
dna_same = dna.loc[same_gene,:]
rna_same = rna.loc[same_gene,:]
```

```python
dna_id = list(dna.columns)
rna_id = list(rna.columns)
same_id = list(set(dna_id).intersection(rna_id))
same_id = [x for x in same_id if x[:5] !="CRC01" and x[:5] !="CRC02"]
same_id
```

```python
dna_same = dna_same.loc[:,same_id]
rna_same = rna_same.loc[:,same_id]
```

```python
dna_same
```

```python
rna_same
```

```python
# delete genes not expressed in all RNA cells

rna_same[rna_same<0] = 0
dna_same[dna_same<0] = 0
# rna_same[rna_same<1] = 1
dna_same[dna_same<1] = 1

rna_zero = rna_same.sum(axis=1)
rna_zero = rna_zero[rna_zero==0]
rna_zero = rna_zero.index.values
len(rna_zero)
```

```python
dna_zero = dna_same.sum(axis=1)
dna_zero = dna_zero[dna_zero==0]
dna_zero = dna_zero.index.values
len(dna_zero)
```

```python
rna_same_nonzero = rna_same.loc[[x for x in list(rna_same.index) if x not in rna_zero]]
dna_same_nonzero = dna_same.loc[[x for x in list(dna_same.index) if x not in rna_zero]]
dna_same_nonzero
```

```python
rna_same_nonzero
```

```python
rna_same.to_csv("rna_same.csv", header=True,index=True)
dna_same.to_csv("dna_same.csv", header=True,index=True)
```

```python
rna_same_nonzero.to_csv("rna_same_nonzero.csv", header=True,index=True)
dna_same_nonzero.to_csv("dna_same_nonzero.csv", header=True,index=True)
```
